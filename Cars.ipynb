```python
import pandas as pd
import matplotlib.pyplot as plt

veriSeti = pd.read_csv('cars.csv')
arabayuksekligi = veriSeti['lug_boot']
guvenlik = veriSeti['safety']

plt.figure(figsize=(10, 5))  # Grafiğin boyutunu ayarlayın
plt.bar(arabayuksekligi , guvenlik)
plt.xlabel('lug_boot')
plt.ylabel('safety')
plt.title('Arabalarda yükseklik ve konfor grafiği')
plt.xticks(rotation=90)
plt.show()
```
